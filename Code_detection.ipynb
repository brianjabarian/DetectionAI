```python
# use 5 ai detector(4 api and one fine-tuned model)

# ä½¿ç”¨5ä¸ªai detecteræ¥è¿›è¡Œai generateå†…å®¹çš„è¯†åˆ«
import json
import torch
import requests
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from tqdm import tqdm

# === Load RoBERTa-based OpenAI Detector ===
MODEL_NAME = "roberta-base-openai-detector"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
detector = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME).eval()

# === Load AI Detectors API keys ===
api_keys = {
    "originality": "ynml4r6b25akd7j9x0iqcg8fusvwoze1",
    "pengram": "78019670-daaf-4b40-9700-32ebe2569644",
    # "zerogpt": "813fef56-8c19-4061-8d5f-7362c9a529c3",
    "gptzero": "30cf8c389f4c4afe9bfcfc729a2ba1ee"
}

# === Local detector
def detect_ai_roberta(text):
    if not text.strip():
        return None
    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=512)
    with torch.no_grad():
        logits = detector(**inputs).logits
    prob_ai = torch.softmax(logits, dim=-1)[0, 1].item()
    return prob_ai

# === API detector wrapper
def detect_ai_external(text, tool):
    if not text.strip():
        return None
    try:
        if tool == "originality":
            headers = {
                "X-OAI-API-KEY": api_keys["originality"],
                "Content-Type": "application/json"
            }
            payload = {
                "check_ai": True,
                "check_plagiarism": False,
                "check_facts": False,
                "check_readability": False,
                "check_grammar": False,
                "check_contentOptimizer": False,
                "storeScan": False,
                "aiModelVersion": "lite",
                "content": text
            }
            response = requests.post("https://api.originality.ai/api/v3/scan", headers=headers, json=payload)
            response.raise_for_status()
            result = response.json()
            ai_flag = (
                result.get("results", {})
                      .get("ai", {})
                      .get("classification", {})
                      .get("AI")
            )
            return 1 if ai_flag == 1 else 0

        elif tool == "pengram":
            headers = {
                "Content-Type": "application/json",
                "x-api-key": api_keys["pengram"]
            }
            payload = {"text": text}
            response = requests.post("https://text.api.pangramlabs.com", headers=headers, json=payload)
            response.raise_for_status()
            result = response.json()

            # èŽ·å–åŽŸå§‹è¿”å›žå€¼
            likelihood = result.get("ai_likelihood")
            prediction = result.get("prediction")

            return {
                "ai_likelihood": likelihood,
                "prediction": prediction
            }

        elif tool == "gptzero":
            headers = {
                "Accept": "application/json",
                "Content-Type": "application/json",
                "x-api-key": api_keys["gptzero"]
            }
            payload = {
                "document": text,
                "multilingual": False
            }

            try:
                response = requests.post(
                    "https://api.gptzero.me/v2/predict/text",
                    headers=headers,
                    json=payload
                )
                response.raise_for_status()
                result = response.json()

                classification = (
                    result.get("documents", [{}])[0]
                          .get("document_classification")
                )

                # âœ… GPTZero classification logic:
                # If document_classification is "AI_ONLY" or "MIXED", return 1
                #     â†’ This means the text is either fully human-written or has significant human characteristics.
                # Else if it's anything else (e.g., "HUMAN_ONLY"), return 0.
                # If classification is missing, return an error message.
                if classification in ["AI_ONLY", "MIXED"]:
                    return 1
                elif classification:
                    return 0
                else:
                    return "Error: document_classification missing"
            except Exception as e:
                return f"Error: {str(e)}"

    except Exception as e:
        return f"Error: {str(e)}"

    except Exception as e:
        print(f"âŒ Error calling {tool} API: {e}")
        return None

# === Load input data
input_path = "/content/drive/MyDrive/detection/pre2020_human_text_by_genre/all_ai_generated_merged.json"
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# === Run detection
for item in tqdm(data, desc="Running detection"):
    human_text = item.get("text", "")
    item["human_verdict"] = {
        "roberta-base-detector": detect_ai_roberta(human_text),
        "originality": detect_ai_external(human_text, "originality"),
        "pengram": detect_ai_external(human_text, "pengram"),
        "zerogpt": detect_ai_external(human_text, "zerogpt"),
        "gptzero": detect_ai_external(human_text, "gptzero"),
    }

    ai_texts = item.get("ai_generated", {})
    item["ai_verdicts"] = {}
    for model_name, text in ai_texts.items():
        item["ai_verdicts"][model_name] = {
            "roberta-base-detector": detect_ai_roberta(text),
            "originality": detect_ai_external(text, "originality"),
            "pengram": detect_ai_external(text, "pengram"),
            "zerogpt": detect_ai_external(text, "zerogpt"),
            "gptzero": detect_ai_external(text, "gptzero"),
        }

# === Save output
output_path = input_path.replace(".json", "_with_multiverdicts.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("âœ… Detection completed. Results saved to:", output_path)
```

```python
# only pengram
import json
import requests
from tqdm import tqdm

# === Load Pangram API key ===
api_keys = {
    "pengram": "78019670-daaf-4b40-9700-32ebe2569644"
}

# === Pangram æ£€æµ‹å‡½æ•° ===
def detect_ai_pangram(text):
    if not text.strip():
        return None
    headers = {
        "Content-Type": "application/json",
        "x-api-key": api_keys["pengram"]
    }
    payload = {"text": text}
    try:
        response = requests.post("https://text.api.pangramlabs.com", headers=headers, json=payload)
        response.raise_for_status()
        result = response.json()

        likelihood = result.get("ai_likelihood")
        prediction = result.get("prediction")

        return {
            # "verdict": 1 if isinstance(likelihood, (float, int)) and likelihood > 0.5 else 0,
            "ai_likelihood": likelihood,
            "prediction": prediction
        }
    except Exception as e:
        return {"error": str(e)}

# === åŠ è½½è¾“å…¥æ•°æ® ===
input_path = "/content/drive/MyDrive/detection/pre2020_human_text_by_genre/merged_corpus_withOpenai_all_detectors.json"
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# === æ‰§è¡Œ Pangram æ£€æµ‹ ===
for item in tqdm(data, desc="Running Pangram detection"):
    human_text = item.get("text", "")
    item["pengram_human_verdict"] = detect_ai_pangram(human_text)

    ai_texts = item.get("ai_generated", {})
    item["pengram_ai_verdicts"] = {}
    for model_name, text in ai_texts.items():
        item["pengram_ai_verdicts"][model_name] = detect_ai_pangram(text)

# === ä¿å­˜è¾“å‡ºç»“æžœ ===
output_path = input_path.replace(".json", "_pangram_only.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("âœ… Pangram-only detection completed. Results saved to:", output_path)
```

```python
# only originality
import json
import requests
from tqdm import tqdm

# === API key for Originality.ai ===
api_keys = {
    "originality": "ynml4r6b25akd7j9x0iqcg8fusvwoze1"
}

# === Originality.ai wrapper that extracts full classification + confidence
def detect_ai_originality(text):
    if not text.strip():
        return None
    try:
        headers = {
            "X-OAI-API-KEY": api_keys["originality"],
            "Content-Type": "application/json"
        }
        payload = {
            "check_ai": True,
            "check_plagiarism": False,
            "check_facts": False,
            "check_readability": False,
            "check_grammar": False,
            "check_contentOptimizer": False,
            "storeScan": False,
            "aiModelVersion": "lite",
            "content": text
        }
        response = requests.post("https://api.originality.ai/api/v3/scan", headers=headers, json=payload)
        response.raise_for_status()
        result = response.json()
        classification = result.get("results", {}).get("ai", {}).get("classification", {})
        confidence = result.get("results", {}).get("ai", {}).get("confidence", {})
        return {
            "classification": {
                "AI": classification.get("AI"),
                "Original": classification.get("Original")
            },
            "confidence": {
                "AI": confidence.get("AI"),
                "Original": confidence.get("Original")
            }
        }
    except Exception as e:
        return {"error": str(e)}

# === Load input data
input_path = "/content/drive/MyDrive/detection/pre2020_human_text_by_genre/merged_corpus_all_genres_generated_openai_withVerdict.json"
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# === Run detection using only Originality.ai
for item in tqdm(data, desc="Running Originality.ai detection"):
    human_text = item.get("text", "")
    item["human_verdict"] = {
        "originality": detect_ai_originality(human_text)
    }

    ai_texts = item.get("ai_generated", {})
    item["ai_verdicts"] = {}
    for model_name, text in ai_texts.items():
        item["ai_verdicts"][model_name] = {
            "originality": detect_ai_originality(text)
        }

# === Save output
output_path = input_path.replace(".json", "_originality_pangram.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("âœ… Originality-only detection completed. Results saved to:", output_path)

```

```python
# only use roBERTa

import json
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from tqdm import tqdm

# === Load RoBERTa-based OpenAI Detector ===
MODEL_NAME = "roberta-base-openai-detector"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
detector = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME).eval()

# === RoBERTa detector: returns probability of AI-generated
def detect_ai_roberta(text):
    if not text.strip():
        return None
    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=512)
    with torch.no_grad():
        logits = detector(**inputs).logits
    prob_ai = torch.softmax(logits, dim=-1)[0, 1].item()
    return prob_ai  # è¿”å›ž AI æ¦‚çŽ‡

# === Load input data
input_path = "/content/drive/MyDrive/detection/pre2020_human_text_by_genre/merged_corpus_all_genres_generated_openai_withVerdict.json"
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# === Run RoBERTa-only detection
for item in tqdm(data, desc="Running RoBERTa detection"):
    human_text = item.get("text", "")
    item.setdefault("human_verdict", {})["roberta-base-detector"] = detect_ai_roberta(human_text)

    ai_texts = item.get("ai_generated", {})
    item.setdefault("ai_verdicts", {})
    for model_name, text in ai_texts.items():
        item["ai_verdicts"].setdefault(model_name, {})["roberta-base-detector"] = detect_ai_roberta(text)

# === Save output
output_path = input_path.replace(".json", "_roberta_only.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print("âœ… RoBERTa-only detection completed. Results saved to:", output_path)
```

```python

# merge all ai-detector, and remain default prob
# Multi-detector AI text analysis with default prob
import json
import requests
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from tqdm import tqdm

# === Load API keys ===
api_keys = {
    "pengram": "78019670-daaf-4b40-9700-32ebe2569644",
    "originality": "ynml4r6b25akd7j9x0iqcg8fusvwoze1",
    "gptzero": "30cf8c389f4c4afe9bfcfc729a2ba1ee"
}

# === Pangram æ£€æµ‹å‡½æ•° ===
def detect_ai_pangram(text):
    if not text.strip():
        return None
    headers = {
        "Content-Type": "application/json",
        "x-api-key": api_keys["pengram"]
    }
    payload = {"text": text}
    try:
        response = requests.post("https://text.api.pangramlabs.com", headers=headers, json=payload)
        response.raise_for_status()
        result = response.json()

        likelihood = result.get("ai_likelihood")
        prediction = result.get("prediction")

        return {
            "ai_likelihood": likelihood,
            "prediction": prediction
        }
    except Exception as e:
        return {"error": str(e)}

# === Originality.ai wrapper ===
def detect_ai_originality(text):
    if not text.strip():
        return None
    try:
        headers = {
            "X-OAI-API-KEY": api_keys["originality"],
            "Content-Type": "application/json"
        }
        payload = {
            "check_ai": True,
            "check_plagiarism": False,
            "check_facts": False,
            "check_readability": False,
            "check_grammar": False,
            "check_contentOptimizer": False,
            "storeScan": False,
            "aiModelVersion": "lite",
            "content": text
        }
        response = requests.post("https://api.originality.ai/api/v3/scan", headers=headers, json=payload)
        response.raise_for_status()
        result = response.json()
        classification = result.get("results", {}).get("ai", {}).get("classification", {})
        confidence = result.get("results", {}).get("ai", {}).get("confidence", {})
        return {
            "classification": {
                "AI": classification.get("AI"),
                "Original": classification.get("Original")
            },
            "confidence": {
                "AI": confidence.get("AI"),
                "Original": confidence.get("Original")
            }
        }
    except Exception as e:
        return {"error": str(e)}

# === GPTZero detector wrapper ===
def detect_ai_gptzero(text):
    if not text.strip():
        return None
    try:
        headers = {
            "Accept": "application/json",
            "Content-Type": "application/json",
            "x-api-key": api_keys["gptzero"]
        }
        payload = {
            "document": text,
            "multilingual": False
        }
        response = requests.post(
            "https://api.gptzero.me/v2/predict/text",
            headers=headers,
            json=payload
        )
        response.raise_for_status()
        doc = response.json().get("documents", [{}])[0]
        return {
            "average_generated_prob": doc.get("average_generated_prob")
        }
    except Exception as e:
        return {"error": str(e)}

# === Load RoBERTa-based OpenAI Detector ===
MODEL_NAME = "roberta-base-openai-detector"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
detector = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME).eval()

# === RoBERTa detector function ===
def detect_ai_roberta(text):
    if not text.strip():
        return None
    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=512)
    with torch.no_grad():
        logits = detector(**inputs).logits
    prob_ai = torch.softmax(logits, dim=-1)[0, 1].item()
    return prob_ai

# === Load input data ===
input_path = "/content/drive/MyDrive/detection/pre2020_human_text_by_genre/merged_corpus_withOpenai.json"
with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

# ç»Ÿè®¡ä¿¡æ¯
total_items = len(data)
skipped_items = 0
processed_items = 0

# === Run all detectors ===
for item in tqdm(data, desc="Running all AI detectors"):
    # æ£€æŸ¥æ˜¯å¦å·²æœ‰æ£€æµ‹ç»“æžœï¼Œå¦‚æžœæœ‰åˆ™è·³è¿‡
    if "human_verdict" in item and "ai_verdicts" in item and all(model in item["ai_verdicts"] for model in item.get("ai_generated", {})):
        skipped_items += 1
        continue

    human_text = item.get("text", "")

    # Initialize verdict structure if needed
    item.setdefault("human_verdict", {})

    # Add results from each detector
    item["human_verdict"]["pengram"] = detect_ai_pangram(human_text)
    item["human_verdict"]["originality"] = detect_ai_originality(human_text)
    item["human_verdict"]["gptzero"] = detect_ai_gptzero(human_text)
    item["human_verdict"]["roberta-base-detector"] = detect_ai_roberta(human_text)

    # Process AI-generated texts
    ai_texts = item.get("ai_generated", {})
    item.setdefault("ai_verdicts", {})

    for model_name, text in ai_texts.items():
        item["ai_verdicts"].setdefault(model_name, {})
        item["ai_verdicts"][model_name]["pengram"] = detect_ai_pangram(text)
        item["ai_verdicts"][model_name]["originality"] = detect_ai_originality(text)
        item["ai_verdicts"][model_name]["gptzero"] = detect_ai_gptzero(text)
        item["ai_verdicts"][model_name]["roberta-base-detector"] = detect_ai_roberta(text)

    processed_items += 1

# === Save output ===
output_path = input_path.replace(".json", "_all_detectors.json")
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"âœ… Multi-detector analysis completed.")
print(f"Total items: {total_items}")
print(f"Skipped items (already processed): {skipped_items}")
print(f"Processed items: {processed_items}")
print(f"Results saved to: {output_path}")
```
